## Contar objetos

```python
#Importar librerías
import cv2
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar imagen
imagen = data.coins()
imagen_cv = imagen.astype('uint8')

#Identificar bordes
bordes = cv2.Canny(imagen_cv, 100, 200)

#Contar los bordes
ctns, _ = cv2.findContours(bordes,
                           cv2.RETR_EXTERNAL,
                           cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(imagen_cv, ctns, -1, (0,0,255), 2)

print('Objetos detectados: ', len(ctns))


# Mostrar resultados
plt.figure(figsize=(6,6))
plt.imshow(imagen_cv, cmap='gray')
plt.title('Conteo de objetos por detección de bordes')
plt.axis('off')
plt.show()
```

```python
#Cargar imagen
imagen = cv2.imread('imagen.JPG')
imagen_cv = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

#Identificar bordes
bordes = cv2.Canny(imagen_cv, 100, 200)

#Contar los bordes
ctns, _ = cv2.findContours(bordes,
                           cv2.RETR_EXTERNAL,
                           cv2.CHAIN_APPROX_SIMPLE)

cv2.drawContours(imagen_cv, ctns, -1, (0,0,255), 2)

print('Objetos detectados: ', len(ctns))


# Mostrar resultados
plt.figure(figsize=(6,6))
plt.imshow(imagen_cv, cmap='gray')
plt.title('Conteo de objetos por detección de bordes')
plt.axis('off')
plt.show()
```
