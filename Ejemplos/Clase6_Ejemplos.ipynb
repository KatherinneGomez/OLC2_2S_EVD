### Modelo de Regresión Lineal

```python
#Importar librerías
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
# Cargar dataset
df = px.data.tips()
df.head()
```

```python
#  Visualizar gráfica para validar el comportamiento de los datos
fig = px.scatter(df, x='total_bill', y='tip', color='sex')
fig.show()
```

```python
# Importar modelo de regresión lineal
from sklearn.linear_model import LinearRegression

# Preparar Datos de entrada
# Variable independiente
total_cuenta = df['total_bill'].values.reshape((-1, 1))

# Crear el modelo
regresion = LinearRegression()

# Entrenar el modelo
modelo = regresion.fit(total_cuenta, df['tip'])

print('Intersección: ', modelo.intercept_)
print('Pendiente: ', modelo.coef_)

# Predecir valores
prueba = modelo.predict([[40]])
print('Predicción: ', prueba)
```

```python
# Guardar la pendiente y la intersección
m = modelo.coef_
b = modelo.intercept_

# Generar los puntos de la línea
x_line = np.linspace(df['total_bill'].min(), df['total_bill'].max(), 100)
y_line = m * x_line + b

# Agregar la línea de regresión
fig.add_trace(
    go.Scatter(
        x=x_line,
        y=y_line,
        mode='lines',
        name='Regresión lineal',
        line=dict(width=3)
    )
)

fig.show()
```

### Modelo de regresión logística
#### Clasificación de correo como SPAM/ NO SPAM


Las variables:
- Número de palabras
- Cantidad de caracteres especiales
- Longitud del mensaje

influyen en si es spam (1) o no spam (0)

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Leer datos
df = pd.read_csv('spam_detection.csv')

# Preparar datos de entrada
x = df[['num_palabras', 'caracteres_especiales', 'longitud_mensaje']]
y = df['es_spam']

# Separar datos en train y test
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42
)

# Entrenar el modelo de Regresión Logística
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

# Hacer predicciones
y_pred = modelo.predict(X_test)
y_pred
```
